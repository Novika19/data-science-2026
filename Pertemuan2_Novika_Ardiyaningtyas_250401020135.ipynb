```python


import pandas as pd
import numpy as np

# 1. Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

# 2. Eksplorasi Awal (Q1 & Q2)
print("Q1 - Shape:", df.shape)  # jumlah baris & kolom
print("Q2 - Missing values:\n", df.isnull().sum())

# 3. Persentase selamat (Q3)
survival_rate = df['Survived'].mean()
print("Q3 - Persentase selamat:", survival_rate)

# 4. Wanita kelas 1 yang selamat (Q4)
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1) & (df['Survived']==1)]
print("Q4 - Rata-rata usia wanita kelas 1 yang selamat:", wk1['Age'].mean())

# 5 & 6. Groupby per kelas (Q5 & Q6)
hasil = df.groupby('Pclass')['Survived'].mean()
print("Q6 - Survival rate per class:\n", hasil)

# Cari kelas dengan survival tertinggi (Q5)
kelas_tertinggi = hasil.idxmax()
print("Q5 - Kelas dengan survival tertinggi:", kelas_tertinggi)
```
