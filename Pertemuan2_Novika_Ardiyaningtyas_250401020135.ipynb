```python
import pandas as pd
import numpy as np

# ① Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

# ② Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())

# ③ Filtering — Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())

# ④ Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','mean'),
).round(3)

print(hasil)



# =========================
# JAWABAN KUIS
# =========================

# Q1
q1_answer = f"{df.shape[0]} baris, {df.shape[1]} kolom"
q1_question = "Q1 Berapa jumlah baris dan kolom dataset Titanic?"

# Q2
q2 = df.isnull().sum()

# Q3
q3_answer = df['Survived'].mean()
q3_question = "Q3 Berapa persen penumpang yang selamat (Survived=1)?"

# Q4
wk1_survived = df[(df['Sex']=='female') & (df['Pclass']==1) & (df['Survived']==1)]
q4_answer = wk1_survived['Age'].mean()
q4_question = "Q4 Rata-rata usia penumpang wanita kelas 1 yang selamat?"

# Q6
q6 = df.groupby('Pclass')['Survived'].mean()

# Q5
q5_answer = q6.idxmax()
q5_question = "Q5 Kelas penumpang mana yang tingkat keselamatan tertinggi?"

# =========================
# TAMPILKAN DALAM TABEL
# =========================

# Tabel ringkasan Q1–Q6
hasil_tabel = pd.DataFrame({
    "Pertanyaan": [q1_question, q3_question, q4_question, q5_question],
    "Jawaban": [
        q1_answer,
        q3_answer,
        q4_answer,
        q5_answer
    ]
})

print("\n===== JAWABAN Q1, Q3, Q4, Q5 ====")
display(hasil_tabel)

# Tabel missing values (Q2)
print("\n===== TABEL MISSING VALUES (Q2) ====")
display(q2.to_frame(name="Jumlah Kosong"))

# Tabel survival per kelas (Q6)
print("\n===== TABEL SURVIVAL PER KELAS (Q6) ====")
display(q6.to_frame(name="Survival Rate"))
```
